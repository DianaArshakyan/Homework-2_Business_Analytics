```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from statsmodels.discrete.discrete_model import Logit
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
```

**Problem 1**

```python
data = pd.read_excel("Employee_attrition.xlsx")
```

```python
data.head()
```

```python
data.info()
```

```python
data_numeric = data.select_dtypes(exclude="object")
data_object = data.select_dtypes(include="object")
```

```python
data_object.head()
```

```python
data_numeric.var()==0
```

```python
data_numeric.EmployeeCount.var()==0
```

```python
# we have two variables ("StandardHouse", "EmployeeCount") that should be dropped, as they have no variance.
```

```python
data_numeric = data_numeric.drop(["StandardHours", "EmployeeCount"], axis = 1)
```

```python
for i in data_object.columns:
    print(data_object[i].unique())
```

```python
# we see that there is an object variable("Over18") with one category("Y"), so we should drop that column.
```

```python
data_object = data_object.drop(["Over18"], axis = 1)
```

```python
data_dummies = pd.get_dummies(data_object, prefix=data_object.columns, drop_first=True)
# we convert object variables into dummy variables 
```

```python
sns.heatmap(data_numeric.corr())
```

```python
plt.figure(figsize=(16, 14)) 
sns.heatmap(data_numeric.corr().abs().round(2), annot=True)
plt.show()
```

```python
data_final = data_numeric.join(data_dummies)
```

```python
# Now we want to see which correlations are above or equal to 75% in absolute values
```

```python
plt.figure(figsize=(12, 10)) 
sns.heatmap(data_numeric.corr().abs()>=0.75)
plt.show()
```

```python
# We see that there are variables that have correlations higher than 75%, so we should delete them
```

```python
del data_numeric["YearsAtCompany"]
```

```python
del data_numeric["TotalWorkingYears"]
```

```python
del data_numeric["JobLevel"]
```

```python
del data_numeric["PercentSalaryHike"]
```

```python
plt.figure(figsize=(12, 10)) 
sns.heatmap(data_numeric.corr().abs()>=0.75)
plt.show()
```

```python
data_final = data_numeric.join(data_dummies)
```

```python
data_final.head()
```

**LOGISTIC REGRESSION**

```python
data_final.Attrition_Yes.value_counts()/len(data_final)*100
```

```python
# We see that about 84% of employees stay in their jobs.
```

```python
Y = data_final.Attrition_Yes
X = data_final.drop("Attrition_Yes",axis=1)
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.25,random_state=42)
```

```python
formula = 'Y~'+'+'.join(X.columns.tolist())
formula
```

```python
logit = Logit.from_formula(formula = formula, data = data_final).fit()
```

```python
logit.summary2()
```

```python
# According to summary of model we can see that there are some statistically significant variables that have 
# p-value less than 0.05. For example, let's interpret the coefficient YearsAtCompany(0.0961):
# One year increase at company will increase employee attrition by 0.0961, ceteris paribus.
```

```python
logit_sk = LogisticRegression()
```

```python
logit_sk = logit_sk.fit(x_train,y_train)
```

```python
logit_sk.score(x_test,y_test)
```

```python
# Our model is accurate by around 88%.
```

**Problem 2**

```python
tree_1 = DecisionTreeClassifier(random_state = 42)
```

```python
tree_1.fit(x_train,y_train)
```

```python
tree_1.score(x_train, y_train)
```

```python
tree_1.score(x_test, y_test)
```

```python
# This result is not desired as it is lower than benchmark.
```

```python
importances = tree_1.feature_importances_
importances = pd.DataFrame(data=importances, index=X.columns, columns=['importances'])
```

```python
importances.sort_values(by = "importances", ascending=False)
```

```python
tree_2 = DecisionTreeClassifier(max_depth = 3, min_samples_leaf = 100, random_state = 42)
tree_2.fit(x_train,y_train)
tree_2.score(x_test,y_test)
```

```python
# Again we see that Logistic Regression gives higher accuracy than Decision tree model.
```
